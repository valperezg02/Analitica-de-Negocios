En este documento se desarrolla y analiza un modelo Naive Bayes para detección de riesgo de lavado de activos (SARLAFT) en una entidad financiera utilizando las variables socioeconómicas de un solicitante de crédito. De acuerdo con lo anterior, las variables socioeconómicas son:

**Edad:** es la edad que posee el cliente al momento del análisis de riesgo.

**Ingresos_usd:** Ingresos recibidos por una persona como resultado de una actividad comercial.

**Gastos_usd:** Son los gastos del cliente. Gastos de personas a cargo, hijos y servicios públicos.

**Num_tarjetas_credito:** es la cantidad de tarjetas de crédito que el cliente posee.

**Monto_transado_tarjetas_usd:** es la cantidad de dinero que se ha movido en las tarjetas.

**Porcentaje_crecimiento_patrimonio:** indica el crecimiento o disminución del patrimonio del cliente.


0. Se cargan las librerías de trabajo

```python
import numpy as np
import pandas as pd

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

1. Se cargan los datos de trabajo de la base de datos LavadoActivos

```python
nxl='/content/2. LavadoActivos.xlsx'
XDB= pd.read_excel(nxl, sheet_name=0)

# Se seleccionan las variables de trabajo
XD=XDB[['edad','ingresos_usd','gastos_usd', 'num_tarjetas_credito', 'monto_transado_tarjetas_usd','porcentaje_crecimiento_patrimonio']]
XD.head(10)

yd=XDB[['lavado_activos']]
yd.head()
```

2. Se implementa el modelo Naive bayes

```python
mnb=GaussianNB()

# se ajustan las variables entrada - salida
mnb.fit(XD,yd)

# Se muestra la media de las variables
m=mnb.theta_

# Se muestra la Desviación estándar de las variables
sigma=mnb.var_ ; sigma=np.sqrt(sigma)
print('edad','ingresos_usd','gastos_usd', 'num_tarjetas_credito', 'monto_transado_tarjetas_usd','porcentaje_crecimiento_patrimonio')
print(m)
print('Las desviaciones son:')
print(sigma)
```

**Análisis de Resultados**

De los resultados arrojados por el modelo, se observa que los clientes riesgosos tienen ingresos mayores a 70 mil USD y los clientes normales tienen ingresos de 6 mil USD aproximadamente. Esto quiere decir que los clientes riesgosos tienen ingresos mucho más grandes.

Pasa lo mismo con los gastos, donde los clientes riesgosos tienen gastos mucho mas altos que los clientes normales. Esto muestra un flujo de grandes cantidades de dinero por parte de los clientes riesgosos que indican una posible existencia de actividades inusuales.

Existe tambien una gran diferencia en la variable Monto transado con tarjetas, ya que los clientes riesgosos transan más de 33 mil USD y lo clientes normales solo alrededor de 3 mil USD. Esta gran diferencia se puede considerar como el indicador principal para reconocer lavado de activos.

Las otras variables (edad, número de tajetas y el porcentaje de crecimiento) tienen valores similares entre los clientes riesgosos y los normales, por lo cual son menos llamativas para el análisis de lavado de activos.
